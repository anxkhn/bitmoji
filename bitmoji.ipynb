```python
import urllib.request
import json
from numpy.random import randint
from io import BytesIO
from PIL import Image
import pandas as pd
```

```python
def display_image_from_url(img_url):
    img_data = urllib.request.urlopen(img_url).read()
    return Image.open(BytesIO(img_data))
def write_file(filename, contents):
    file = open(filename, "w")
    file.write(contents)
    file.close()
```

```python
json_url = "https://api.bitmoji.com/content/templates?app_name=bitmoji"
json_data = urllib.request.urlopen(json_url).read()
raw_data = json.loads(json_data.decode())
```

```python
rohit_id = "4caac462-971c-47f5-8f78-0978bd5fa213"
rooba_id = "b84252ec-d02e-4a50-8682-0445ff64379c"
```

```python
raw_data.keys()
```

```python
num_solo_comics = len(raw_data["imoji"])
num_partner_comics = len(raw_data["friends"])
print(num_solo_comics, num_partner_comics)
```

## Solo Comics

```python
user_id = rohit_id
img_html_list = ['<img src="%s" width=100/>' % raw_data["imoji"][i]["src"] % user_id for i in range(num_solo_comics)]
write_file("rohit.html", "".join(img_html_list))
```

```python
user_id = rooba_id
img_html_list = ['<img src="%s" width=100/>' % raw_data["imoji"][i]["src"] % user_id for i in range(num_solo_comics)]
write_file("rooba.html", "".join(img_html_list))
```

## Solo Poses

```python
user_id = rohit_id
img_html_list = ['<img src="https://render.bitstrips.com//render/%s/%s-v1.png" width=100/>' % (raw_data["imoji"][i]["comic_id"], user_id) for i in range(num_solo_comics)]
write_file("rohit_poses.html", "".join(img_html_list))
```

```python
user_id = rooba_id
img_html_list = ['<img src="https://render.bitstrips.com//render/%s/%s-v1.png" width=100/>' % (raw_data["imoji"][i]["comic_id"], user_id) for i in range(num_solo_comics)]
write_file("rooba_poses.html", "".join(img_html_list))
```

## Partner Comics

```python
first_id = rooba_id
second_id = rohit_id
img_html_list = ['<img src="%s" width=100/>' % raw_data["friends"][i]["src"] % (first_id, second_id) for i in range(num_partner_comics)]
write_file("rooba_rohit.html", "".join(img_html_list))
```

```python
first_id = rohit_id
second_id = rooba_id
img_html_list = ['<img src="%s" width=100/>' % raw_data["friends"][i]["src"] % (first_id, second_id) for i in range(num_partner_comics)]
write_file("rohit_rooba.html", "".join(img_html_list))
```
